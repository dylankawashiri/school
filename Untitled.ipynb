```python
import numpy as np
from matplotlib import pyplot as plt
import pandas as pd
```

```python
url = "https://philchodrow.github.io/PIC16A/datasets/palmer_penguins.csv"
penguins = pd.read_csv(url)

penguins.head()
```

```python
from sklearn.model_selection import train_test_split

np.random.seed(3354354524)
train, test = train_test_split(penguins, test_size = 0.2)
```

```python
from sklearn import preprocessing
def prep_penguins_data(data_df):
    df = data_df.copy()
    le = preprocessing.LabelEncoder()
    df['Sex'] = le.fit_transform(df['Sex'])
    df['Island'] = le.fit_transform(df['Island'])
    df['Species'] = le.fit_transform(df['Species'])
    
    df = df.drop(['Comments'], axis = 1)
    df = df.drop(['studyName'], axis = 1)
    df = df.drop(['Sample Number'], axis = 1)
    df = df.drop(['Stage'], axis = 1)
    df = df.drop(['Individual ID'], axis = 1)
    df = df.drop(['Clutch Completion'], axis = 1)
    df = df.drop(['Date Egg'], axis = 1)
    df = df.dropna()
    X = df.drop(['Species'], axis = 1)
    y = df['Species']
        
    return(X, y)

X_train, y_train = prep_penguins_data(train)
X_test,  y_test  = prep_penguins_data(test)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import cross_val_score

LR=LogisticRegression()
cross_val_score(LR,X_train,y_train,cv=5).mean()
```
